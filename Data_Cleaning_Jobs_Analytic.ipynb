```python
import pandas as pd
```

```python
d1 = pd.read_csv('/content/sc.csv')
d2 = pd.read_csv('/content/scrap.csv')
d3 = pd.read_csv('/content/scrap_data.csv')
d4 = pd.read_csv('/content/scrapy_data.csv')
d5 = pd.read_csv('/content/scrape.csv')
d6 = pd.read_csv('/content/sca.csv')
d7 = pd.read_csv('/content/scar.csv')
```

```python
df = pd.concat([d1,d2,d3,d4,d5,d6,d7],ignore_index =True)
```

```python
df.rename(columns={'comapany': 'name'}, inplace=True)
df.rename(columns={'employeecount': 'employees_count'}, inplace=True)
df.rename(columns={'followers': 'linkedin_followers'}, inplace=True)
df.rename(columns={'totalapplicants': 'total_applicants'}, inplace=True)
```

```python
df.columns
```

```python
df['job_id'] = range(1, len(df) + 1)
df['company_id'] = range(1, len(df) + 1)
df['details_id'] = range(1, len(df) + 1)
```

```python
columns_order = ['job_id', 'company_id', 'details_id'] + [col for col in df.columns if col not in ['job_id', 'company_id', 'details_id']]
df = df[columns_order]
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['linkedin_followers'] = df['linkedin_followers'].str.replace(',','')
df['linkedin_followers'] = df['linkedin_followers'].str.extract(r'(\d+)')
```

```python
df['linkedin_followers'] = df['linkedin_followers'].fillna(0)
df['linkedin_followers'] = df['linkedin_followers'].astype(int)
df['linkedin_followers'] = df['linkedin_followers'].apply(lambda x: f"{x:,}")
```

```python
df['linkedin_followers']
```

```python
df['employees_count'] = df['employees_count'].str.replace(' employees', '')
```

```python
df['employees_count']
```

```python
df['total_applicants'] = df['total_applicants'].str.extract(r'(\d+)\s+applicants' ,expand=False)
df['total_applicants'] = df['total_applicants'].where(df['total_applicants'].notna(), other='')
```

```python
df['level']
```

```python
df['industry'] = df['industry'].str.extract(r'^([^\d]+)',expand=False).str.strip()
```

```python
df['industry']
```

```python
df
```

```python
def clean_level(value):
    if isinstance(value, str):
        if 'Full-time' in value:
            return 'Full-Time'
        elif 'Internship' in value:
            return 'Internship'
        elif 'Mid-Senior level' in value:
            return 'Mid-Senior level'
        elif 'Onsite' in value:
          return 'Onsite'
        elif 'Entry level' in value:
            return 'Entry level'
    return value

df['level'] = df['level'].apply(clean_level)
```

```python
df['level'] = df['level'].str.extract(r'(\w+)')
```
